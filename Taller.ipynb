```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return x**3 - 3*x**2 + x - 1

# Crear un rango de valores para x
x = np.linspace(-2, 3, 400)
y = f(x)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^3 - 3x^2 + x - 1', color='blue')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.title('Gráfica de y = x^3 - 3x^2 + x - 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
def f_prime(x):
    return 3*x**2 - 6*x + 1
```

```python
from scipy.optimize import newton
points = []

def f_with_logging(x):
    points.append(x) 
    return f(x)  

x0 = 100

x_r = newton(f_with_logging, x0, fprime=f_prime)

print("Raíz encontrada:", x_r)

print("\nEstimaciones visitadas:")
for point in points:
    print(point)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y, label="$y = x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axhline(0, color="black", linewidth=0.5)
plt.scatter(x_r, f(x_r), color="red", zorder=5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show
```
